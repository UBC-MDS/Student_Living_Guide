```python
import pandas as pd
import seaborn as sns
```

```python
coli_df = pd.read_csv("data/processed_data.csv")
```

```python
coli_df.head()
```

```python
# Scatter plot between Rent Index and Cost of Living Index
scatter_plot = (sns.scatterplot(data=coli_df, x='Rent Index', y='Cost of Living Index').
 set(title='Scatter Plot: Rent Index vs Cost of Living Index'))
```

```python
# Distribution plot of Restaurant Price Index
distribution_plot = (sns.histplot(data=coli_df, x='Restaurant Price Index', kde=True).
 set(title = 'Distribution Plot: Restaurant Price Index'))
```

```python
# Box plot of Grocery Price Index by Continent
box_plot = (sns.boxplot(data=coli_df, x='Continent', y='Groceries Index').
 set(title = 'Box Plot: Grocery Price Index by Continent'))
```

```python
# Pair plot of different factors against Cost of Living Index
pairwise_plot = (sns.pairplot(data=coli_df, 
                    vars=['Cost of Living Index', 'Rent Index', 'Groceries Index', 'Restaurant Price Index', 'Local Purchasing Power Index']))
```
